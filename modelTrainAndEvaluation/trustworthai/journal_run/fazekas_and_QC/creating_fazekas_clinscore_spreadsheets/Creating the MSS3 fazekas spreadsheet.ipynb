### Creating the MSS3 INTER RATER Fazekas spreadsheet

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
from trustworthai.journal_run.new_MIA_fazekas_and_QC.creating_fazekas_clinscore_spreadsheets.spreadsheet_helper_funcs import *
```

```python
sample_div_and_metrics = "/home/s2208943/preprocessed_data/MSS3_InterRaterData/feature_spreadsheets/ssn_ens_sample_div_and_metrics.csv"
sample_div_and_metrics = pd.read_csv(sample_div_and_metrics)
sample_div_and_metrics = sample_div_and_metrics[[v.endswith("V1") or v.endswith("V4") for v in sample_div_and_metrics['ID'].values]]

```

```python
sample_div_and_metrics['ID'].values
```

```python
spreadsheet_file = "/home/s2208943/preprocessed_data/MSS3_InterRaterData/MVH_JPhitidis_MSS3PtData_20230512.xlsx"
```

```python
fazekas = pd.read_excel(spreadsheet_file)
```

```python
fazekas
```

```python
deep_column = []
pv_column = []
total_column = []

for ID in sample_div_and_metrics['ID'].values:
    subset = fazekas[fazekas['MSS3_Study_Number'] == ID[:-3]]
    if ID.endswith("V1"):
        deep_column.append(subset['V1_DWMLOverall'].values[0])
        pv_column.append(subset['V1_PVLOverall'].values[0])
        
    elif ID.endswith("V4"):
        deep_column.append(subset['V4_DWMLOverall'].values[0])
        pv_column.append(subset['V4_PVLOverall'].values[0])
        
    else:
        raise ValueError
        
total_column = np.array(deep_column) + np.array(pv_column)
```

```python
MSS3_clinscore_df = pd.DataFrame({
    "Patient ID":sample_div_and_metrics['ID'].values,
    "WMH_PV":pv_column,
    "WMH_Deep":deep_column,
    "Total":total_column
    
})
```

```python
MSS3_clinscore_df.to_csv("/home/s2208943/preprocessed_data/MSS3_InterRaterData/clinscore_data.csv")
```
