### preprocess the spreadsheet

- rename the columns
- add a total fazekas column
- add a domain column (called protocol, and one hot encoded)
- remap the subject ids
- save to csv

```python
fazekas.keys()
```

```python
fazekas['Protocol'] = fazekas['SubjectID'].apply(lambda s : s.split("_")[0])
fazekas = one_hot_encode(fazekas, 'Protocol')
```

```python
fazekas = fazekas.rename(columns={'SubjectID':'Patient ID', 'Fazekas_PV':'WMH_PV', 'Fazekas_Deep':'WMH_Deep'})
```

```python
fazekas['Total'] = fazekas['WMH_PV'] + fazekas['WMH_Deep'] 
```

```python
subject_ids_map = {
    "UMC":"Utrecht",
    "SIN":"Singapore",
    "AMS":"Amsterdam_GE3T",
}
```

```python
fazekas['Patient ID'] = [subject_ids_map[idx.split("_")[0]] + "_" + idx.split("_")[1] for idx in fazekas['Patient ID'].values]
```
