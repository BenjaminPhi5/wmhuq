```python
# augmentation definintion
def get_transforms(is_3D):
    transforms = [
        LabelSelect(label_id=1),
        RandomFlip(p=0.5, orientation="horizontal"),
        # GaussianBlur(p=0.5, kernel_size=7, sigma=(.1, 1.5)),
        # GaussianNoise(p=0.2, mean=0, sigma=0.2),
        # RandomAffine(p=0.2, shear=(.1,3.)),
        # RandomAffine(p=0.2, degrees=5),
        #RandomResizeCrop(p=1., scale=(0.6, 1.), ratio=(3./4., 4./3.))
        RandomResizeCrop(p=1., scale=(0.3, 0.5), ratio=(3./4., 4./3.)) # ssn
    ]
    if not is_3D:
        transforms.append(lambda x, y: (x, y.squeeze().type(torch.long)))
        return PairedCompose(transforms)
    else:
        transforms.append(CropZDim(size=32, minimum=0, maximum=-1))
        transforms.append(lambda x, y: (x, y.squeeze().type(torch.long)))
        return PairedCompose(transforms)
```

```python
# function to do train validate test split
test_proportion = 0.1
validation_proportion = 0.2

def train_val_test_split(dataset, val_prop, test_prop, seed):
    # I think the sklearn version might be prefereable for determinism and things
    # but that involves fiddling with the dataset implementation I think....
    size = len(dataset)
    test_size = int(test_prop*size) 
    val_size = int(val_prop*size)
    train_size = size - val_size - test_size
    train, val, test = random_split(dataset, [train_size, val_size, test_size], generator=torch.Generator().manual_seed(seed))
    return train, val, test
```

```python
def actual_dice(y_hat, y_true):
    y_hat = torch.nn.functional.softmax(y_hat, dim=1).argmax(dim=1)
    # count how many times the ground truth and target have nothing in them
    s0 = y_hat.shape[0]
    y_hat = y_hat.view(s0, -1)
    y_true = y_true.view(s0, -1)
    # print(torch.sum(torch.sum(y_true, dim=1) == 0))
    # print(torch.sum(torch.sum(y_hat, dim=1) == 0))
    # print(torch.sum(torch.logical_and(torch.sum(y_hat, dim=1) == 0, torch.sum(y_true, dim=1) == 0)))
    ignores = torch.sum(torch.logical_and(torch.sum(y_hat, dim=1) == 0, torch.sum(y_true, dim=1) == 0))
    
    
    numerator = torch.sum(2 * y_true * y_hat, dim=1)
    # print(numerator)
    denominator = torch.sum(y_true + y_hat, dim=1)
    # print(denominator)
    
    return torch.sum((numerator)/ (denominator + 1e-8)) / (s0-ignores) # s0
```

```python
def slice_dice(y_hat, y_true):
    # print(y_hat.shape)
    y_hat = torch.nn.functional.softmax(y_hat, dim=0).argmax(dim=0)
    
    if torch.sum(y_hat) == 0 and torch.sum(y_true) == 0:
        return "n/a"
    
    # print(y_hat.shape)
    # print(y_true.shape)
    
    numerator = torch.sum(2 * y_hat * y_true)
    denominator = torch.sum(y_hat + y_true)
    return numerator / (denominator + 1e-8)
```

```python
def dice(y_pred, y_true):
    y_pred = torch.nn.functional.softmax(y_pred, dim=1).argmax(dim=1)
    denominator = torch.sum(y_pred) + torch.sum(y_true)
    numerator = 2. * torch.sum(torch.logical_and(y_pred, y_true))
    return numerator / denominator
```

### Building the patch uncertainty DDU map

```python
def get_density_mapper():
    return nn.Sequential(
        nn.Conv2d(2, 8, 2),
        nn.Conv2d(8, 8, 2),
        nn.Conv2d(8, 4, 2),
        nn.Flatten()
    )
```

```python
summary(get_density_mapper(), (1, 2, 4, 4))
```

```python
class GMM_model_arbitrary(nn.Module):
    def __init__(self, model, device='cpu'):
        super().__init__()
        window = 4
        self.window = window
        self.get_model = lambda : model.to(device) # wrapping in the lambda stops the model from being a saveable parameter
        self.dists = (window ** 2) + 1
        self.event_shape = (window // 2) ** 2
        self.mu = nn.parameter.Parameter(torch.randn(self.dists, self.event_shape))
        self.cov_l = nn.parameter.Parameter(torch.randn(self.dists, int(0.5 * self.event_shape * (self.event_shape + 1))))
        self.dist_probs = nn.parameter.Parameter(torch.randn(self.dists))
        self.mappers = nn.ModuleList([get_density_mapper() for d in range(self.dists)])
        self.device=device
        
    def forward(self, x):
        # get model prediction
        with torch.no_grad():
            model = self.get_model()
            model.eval()
            device=self.device
            x = x.to(device)
            y_hat = model(x)
            #y_hat = torch.nn.functional.softmax(y_hat, dim=1)
            
        # load model states
        mappers = self.mappers.to(device)
        mus = self.mu.to(device)
        dist_probs = torch.nn.functional.softmax(self.dist_probs, dim=0)
        covls = [torch.zeros(self.event_shape, self.event_shape).to(device) for c in range(self.dists)]
        tril_indices = torch.tril_indices(row=self.event_shape,col=self.event_shape,offset=0).to(device)
        for c in range(self.dists):
            covls[c][tril_indices[0], tril_indices[1]] = self.cov_l[c].to(device)
        covs = [covls[c] @ covls[c].T for c in range(self.dists)]
        
        # generate the multivariate normals
        gmm_normals = [td.MultivariateNormal(mus[c], covs[c]) for c in range(self.dists)]
        
        # get mask
        mask = x[:,1]
        
        return (y_hat, mask, mappers, gmm_normals, dist_probs)
    
    def get_density(self, x):
        print("THIS FUNCTION IS FOR A SINGLE IMAGE ONLY, NOT A BATCH!!")
        with torch.no_grad():
            window = self.window
            y_hat, mask, mappers, gmm_normals, dist_probs = self(x.unsqueeze(0))
            
            all_patches = y_hat.squeeze().unfold(1, window, 1).unfold(2,window,1).unfold(3, window, 1).reshape(2, -1, 4, 4).swapaxes(0,1)
            
            region_probs = [
             0.9749227980358862,
             0.0015754791604434344,
             0.0024972121613805317,
             0.0013054011677627543,
             0.0031901674068228747,
             0.0005098923797722359,
             0.001585615435252076,
             0.0005728011798740052,
             0.0023268123782793863,
             0.0005346667133303196,
             0.0009848160636430265,
             0.00040179147765829977,
             0.001955371637512291,
             0.0004200600073277425,
             0.0008504305520682866,
             0.0005117221371159161,
             0.005854962105870631
            ]
            
            densities = [
                gmm_normals[m].log_prob(
                    gmm.mappers[m](all_patches)
                ) * region_probs[m]
                for m in range(len(gmm.mappers))
            ]
            
            densities = torch.stack(densities).exp()
            max_densities = torch.max(densities, dim=0)[0]
            
            print(max_densities.shape)
            return densities
            
            
    
    def loss(self, gmm_out, target):
        
        window = self.window
        y_hat, mask, mappers, gmm_normals, dist_probs = gmm_out
        y = target
        total_loss = 0
        total_patches = 0
        patch_pixels = window ** 2
        
        # loop over each patch
        for xi in range(0, y.shape[-2] + 1 - window):
            for yi in range(0, y.shape[-1] + 1 - window):
                #print("patch: ", xi, yi)
                # get the patch and the lesion proportion
                patch_y = y[:, xi:xi+window, yi:yi+window]
                patch_p = y_hat[:, :, xi:xi+window, yi:yi+window]
                patch_brain = mask[:, xi:xi+window, yi:yi+window]
                
                # print("patches shape: ", patch_y.shape, patch_p.shape, patch_brain.shape)
                
                lesion_voxels = torch.sum(patch_y == 1, dim=(1,2))
                is_brain = torch.sum(patch_brain >0.1, dim=(1,2))

                brain_proportion = is_brain / patch_pixels
                lesion_proportion = lesion_voxels / patch_pixels
                
                # print(": masking stuff shapes", lesion_voxels.shape, is_brain.shape)

                where_brain = torch.where(brain_proportion > 0.9)[0]
                masked_proportions = lesion_proportion[where_brain]
                masked_patches = patch_p[where_brain]
                
                # print("patches shape : ", patch_p.shape)
                # print("masked patches shape: ",  masked_patches.shape)
                
                # per patch compute the mapping under the relevant mapper function using y
                
                mapper_ids = torch.round(masked_proportions * patch_pixels / (patch_pixels / self.dists))
                # print(": unique mapper ids seen", mapper_ids.unique())
                mapper_outs = []
                
                for m, mapper in enumerate(mappers):
                    m_patches = masked_patches[mapper_ids == m]
                    if len(m_patches) == 0:
                        continue
                    mapper_outs = mapper(m_patches)
                    # print(": m_patches shape", m_patches.shape)
                    # print(": mapper_outs shape", mapper_outs.shape)
                    densities = gmm_normals[m].log_prob(mapper_outs)
                    total_loss += torch.sum(densities)
                    
                total_patches += masked_proportions.shape[0]
                # print(masked_proportions.shape[0])
        
        loss = total_loss / total_patches
        
        return -loss
    
    def non_overlapping_loss(self, gmm_out, target):
        
        window = self.window
        y_hat, mask, mappers, gmm_normals, dist_probs = gmm_out
        y = target
        total_loss = 0
        total_patches = 0
        patch_pixels = window ** 2
        
        # loop over each patch
        patch_y = y.unfold(1, window, window).unfold(2,window,window).reshape(-1,window,window)
        classes = y_hat.shape[1]
        patch_p = y_hat.unfold(2, window, window).unfold(3,window,window).permute(0,2,3,1,4,5).reshape(-1,classes,window,window)
        patch_brain = mask.unfold(1, window, window).unfold(2,window,window).reshape(-1,window,window)
                
        # print("patches shape: ", patch_y.shape, patch_p.shape, patch_brain.shape)
                
        lesion_voxels = torch.sum(patch_y == 1, dim=(1,2))
        is_brain = torch.sum(patch_brain >0.1, dim=(1,2))

        brain_proportion = is_brain / patch_pixels
        lesion_proportion = lesion_voxels / patch_pixels

        # print(": masking stuff shapes", lesion_voxels.shape, is_brain.shape)

        where_brain = torch.where(brain_proportion > 0.9)[0]
        masked_proportions = lesion_proportion[where_brain]
        masked_patches = patch_p[where_brain]

        # print("patches shape : ", patch_p.shape)
        # print("masked patches shape: ",  masked_patches.shape)

        # per patch compute the mapping under the relevant mapper function using y

        # minus 1 because number of dists includes the 0 patches distribution
        mapper_ids = torch.round(masked_proportions * patch_pixels / (patch_pixels / (self.dists-1)))
        # print(torch.unique(masked_proportions * patch_pixels / (patch_pixels / (self.dists-1))))
        # print(": unique mapper ids seen", mapper_ids.unique())
        mapper_outs = []

        for m, mapper in enumerate(mappers):
            m_patches = masked_patches[mapper_ids == m]
            if len(m_patches) == 0:
                continue
            mapper_outs = mapper(m_patches)
            # print(": m_patches shape", m_patches.shape)
            # print(": mapper_outs shape", mapper_outs.shape)
            densities = gmm_normals[m].log_prob(mapper_outs)
            # print(mapper_outs.shape)
            # print(densities.shape)
            total_loss += torch.sum(densities)

        total_patches += masked_proportions.shape[0]
        # print(masked_proportions.shape[0])
        
        loss = total_loss / total_patches
        
        return -loss
```

```python
# compute the mus
with torch.no_grad():
    classes = 2
    mus = torch.zeros(classes, classes)
    pi_s = torch.zeros(classes)
    num_voxels = 0
    for i, (x, y) in tqdm(enumerate(train_dataloader), total=len(train_dataloader), ncols=150, position=0, leave=True):
        y_hat = model(x)
        
        for c in range(classes):
            c_map = (y == c).expand(2, *y.shape).swapaxes(0,1) # make a shape of [n, c, dims...]
            y_hat_at_c = y_hat * c_map
            total = torch.sum(y_hat_at_c, dim=[0,2,3])
            elems = torch.sum(y==c)
            mus[c] +=  (1./(i+1.)) * (total/ elems - mus[c])  # average update rule for mu's
            pi_s[c] += elems
            num_voxels += elems
            
    pi_s /= num_voxels
```

```python
def GMM_density_learnable(zs, gmm):
    # zs is of shape <some dims.., c> note c is the last dim
    # we then put c at the front for the conditoinal densities array
    conditional_densities = torch.zeros(classes, *zs.shape[:-1])
    for c in range(classes):
        conditional_densities[c] = gmm.normals[c].log_prob(zs) + pis[c]
    log_density = torch.sum(conditional_densities, dim=0)
    return log_density

def batch_GMM_density_learnable(y_hat):
    # y hat is of shape <n, c, dims...>
    y_hat = y_hat.permute(0, *list(range(2,len(y_hat.shape))), 1)
    print(y_hat.shape)
    return GMM_density(y_hat)
```

```python
import torch.nn as nn
class GMM_model(nn.Module):
    def __init__(self, ):
        super().__init__()
        self.mu = nn.parameter.Parameter(torch.randn(classes, classes))
        self.cov_l = nn.parameter.Parameter(torch.randn(classes, int(0.5 * classes * (classes + 1))))
        
    def forward(self, x):
        with torch.no_grad():
            model.eval()
            device = model.device
            y_hat = model(x.to(device))
            y_hat = torch.nn.functional.softmax(y_hat, dim=1)
        mus = self.mu
        covls = [torch.zeros(classes, classes).to(device) for c in range(classes)]
        tril_indices = torch.tril_indices(row=classes,col=classes,offset=0).to(device)
        # print(tril_indices.device)
        # print(covls[0].device)
        # print(self.cov_l.device)
        # print(device)
        for c in range(classes):
            covls[c][tril_indices[0], tril_indices[1]] = self.cov_l[c]
        covs = [covls[c] @ covls[c].T for c in range(classes)]
        
        gmm_normals = [td.MultivariateNormal(mus[c], covs[c]) for c in range(classes)]
        
        return (y_hat, gmm_normals)
    
    def get_density(self, x):
        y_hat, gmm_normals = self(x)
        y_hat = y_hat.permute(0, *list(range(2,len(y_hat.shape))), 1)
        conditional_densities = torch.zeros(classes, *y_hat.shape[:-1])
        for c in range(classes):
            conditional_densities[c] = gmm_normals[c].log_prob(y_hat) + pis[c]
        log_density = torch.sum(conditional_densities, dim=0)
        
        return log_density
    
    def loss(self, gmm_out, target):
        y_hat, gmm_normals = gmm_out
        y_hat = y_hat.permute(0, *list(range(2,len(y_hat.shape))), 1).reshape(-1,2)
        y = target.view(-1)
        #print(y_hat.shape, y.shape)
        log_density_sum = torch.Tensor([0.])
        
        for c in range(classes):
            locs = y == c
            voxels_c = y_hat[locs,:]
            conditional_densities = gmm_normals[c].log_prob(voxels_c) + pis[c]
            log_density_sum += torch.sum(conditional_densities, dim=0)
        
        return -log_density_sum / y.shape[0]
```

```python
class GMM_model_arbitrary(nn.Module):
    def __init__(self, ):
        super().__init__()
        self.dists = 10
        self.mu = nn.parameter.Parameter(torch.randn(self.dists, classes))
        self.cov_l = nn.parameter.Parameter(torch.randn(self.dists, int(0.5 * classes * (classes + 1))))
        self.dist_probs = nn.parameter.Parameter(torch.randn(self.dists))
        
    def forward(self, x):
        with torch.no_grad():
            model.eval()
            device = model.device
            y_hat = model(x.to(device))
            #y_hat = torch.nn.functional.softmax(y_hat, dim=1)
        mus = self.mu
        covls = [torch.zeros(classes, classes).to(device) for c in range(self.dists)]
        tril_indices = torch.tril_indices(row=classes,col=classes,offset=0).to(device)
        # print(tril_indices.device)
        # print(covls[0].device)
        # print(self.cov_l.device)
        # print(device)
        for c in range(self.dists):
            covls[c][tril_indices[0], tril_indices[1]] = self.cov_l[c]
        covs = [covls[c] @ covls[c].T for c in range(self.dists)]
        
        gmm_normals = [td.MultivariateNormal(mus[c], covs[c]) for c in range(self.dists)]
        
        dist_probs = torch.nn.functional.softmax(self.dist_probs, dim=0)
        
        return (y_hat, gmm_normals, dist_probs)
    
    def get_density(self, x):
        y_hat, gmm_normals, dist_probs = self(x)
        y_hat = y_hat.permute(0, *list(range(2,len(y_hat.shape))), 1)
        conditional_densities = torch.zeros(self.dists, *y_hat.shape[:-1])
        for c in range(self.dists):
            conditional_densities[c] = gmm_normals[c].log_prob(y_hat) + dist_probs[c]
        log_density = torch.sum(conditional_densities, dim=0)
        
        return log_density
    
    def loss(self, gmm_out, target):
        y_hat, gmm_normals, dist_probs = self(x)
        y_hat = y_hat.permute(0, *list(range(2,len(y_hat.shape))), 1)
        conditional_densities = torch.zeros(self.dists, *y_hat.shape[:-1])
        for c in range(self.dists):
            conditional_densities[c] = gmm_normals[c].log_prob(y_hat) + dist_probs[c]
        log_density = torch.sum(conditional_densities, dim=0)
        
        return -torch.mean(log_density)
```
